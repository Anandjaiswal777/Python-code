Polymorphism is one of the most important concepts in OOP.

```python
# Define a class named car
class Car:
  # Constructor to initialize object attributes
  def__init__(self, brand, model, year):
    self.brand = brand # attribute brand of the car
    self.model = model # atttribute model of the car
    self.year = year # attribute year of the car
  # method to display information about the car
  def car_info(self):
    print(f"This car is a {self.year} {self.brand} {self.model}")
  # method to simulate starting the car
  def start(self):
    print(f"The {self.brand} {self.model} {self.year} is now starting.")
  # method to simulate stopping the car
  def stop(self):
    print(f"The {self.brand} {self.model} {self.year} is now stopping.")
# create an obect (instance) of the car class
my_car = Car("Toyota","Corolla",2020)
# access attributes and call methods
my_car.car_info()
my_car.start()
my_car.stop()
```

```python
car1 = Car("Honda","Civic",2018)
car2 = Car("Ford","Mustang",2022)

car1.car_info() #output: This car is a 2018 Honda Civic
car2.car_info() #output: This car is a 2022 Ford Mustang
```

Single inhertance

```python
# Parent Class
class Animal:
  def speak(self):
    print("Animal speak")

# Child Class
class Dog(Animal):
  def bark(self):
    print("Dog barks")

# Create object of Dog class
dog = Dog()
dog.speak()  #Inherited method
dog.bark()  #Method from dog class
```

Multiple inhertance

```python
# Parent class 1
class Father:
  def walk(self):
    print("Father is walking")

# Parent class 2
class Mother:
  def talk(self):
    print("Moither is talking")

# Child class inherits from both Father and Mother
class Child(Father,Mother):
  def play(self):
    print("Child is playing")

# Create object of Child class
child = Child()
child.walk() #Inherited from Father
child.talk() #Inherited from Mother
child.play() #Child's own method
```

Multi Level Inherittance

```python
# Grandparent class
class Animal:
  def eat(self):
    print("Animal eats")

# Parent class inherits from Animal
class Mammal(Animal):
  def walf(self):
    print("Mammal walk")

# Child class inherits from Mammal
class Dog(Mammal):
  def bark(self):
    print("Dog barks")

# Create objects of Dog class
dog = Dog()
dog.eat() #inherited from animal
dog.walf()  #inherted from mammal
dog.bark() #Dog's own method
```

Hierarchical Inheritance

```python
# Parent class
class Animal:
  def speak(self):
    print("Animal speak")

# Child class 1
class Dog(Animal):
  def bark(self):
    print("Dog barks")

# Child class 2
class Cat(Animal):
  def meow(self):
    print("Cat meows")

# Create objects of Dog and Cat classes
dog = Dog()
cat = Cat()

dog.speak()  #Inherited from Animal
dog.bark()   #Method from Dog class

cat.speak()  #Inherited from Animal
cat.meow()  #Method from Cat class
```

Abstraction

```python
from abc import ABC, abstractmethod

# Abstract class
class Animal(ABC):
  @abstractmethod
  def sound(self):
    pass #Abstract method (no implementation)

# Concrete subclass that implements the abstract method
class Dog(Animal):
  def sound(self):
    return "Bark"

# Another concrete subclass
class Cat(Animal):
  def sound(self):
    return "Meow"

# instantiate objects of the subclasses
dog = Dog()
cat = Cat()

# Call the implemented abstract methods
print(dog.sound())  #Output: Bark
print(cat.sound())  #Output: Meow
```

 Polymorphism

```python
class Dog:
  def sound(self):
    return "Bark"

class Cat:
  def sound(self):
    return "Meow"

# Polymorphism different classes with the same method name
def animal_sound(animal):
  print(animal.sound())

# Create objects of both classes
dog = Dog()
cat = Cat()

# Call the polymorphic function with different objects
animal_sound(dog)  #Output: Bark
animal_sound(cat)  #Output: Meow
```

Encapsulation

```python
class Person:
  def __int__(self,name,age):
    self.__name = name  # private attribute
    self.__age = age  # private attribute
  def get_name(self):
    return self.__name
  def get_age(self):
    return self.__age
  def set_age(self):
    return self.__age
  def set_age(self,age):
    if age > 0:
      self.__age = age
    else:
      print("Age must be positive")

# Usage
person = Person("Alice", 30)
print(person.get_name())  #Accessing the private attribute via getter
print(person.get_age())   #Accessing the private attribute via getter

person.set_age(35)   #Setting the private attribute via setter
print(person.get_age())
```
